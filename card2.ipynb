```python
import pandas as pd
import os
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#MAC font setting for mpl
mpl.rcParams['axes.unicode_minus'] = False 
mpl.rcParams['font.family'] = "AppleGothic"
```

```python
path = os.path.join("card_data_final_2.xlsx")
path
```

```python
df = pd.read_excel(path, sheet_name=0)
df
```

```python
#df_loc.info()
```

```python
df.loc[:,'사용 건수'] = 1
```

```python
b_grouped_count = df['사용 건수'].groupby(df['업종'])
grouped_count = b_grouped_count.sum()
category_count = grouped_count.to_frame()
category_count = category_count.sort_values(by=['사용 건수'], ascending=False)
category_count
```

```python
b_grouped_amount = df['사용 금액'].groupby(df['업종'])
grouped_amount = b_grouped_amount.sum()
category_amount = grouped_amount.to_frame()
category_amount = category_amount.sort_values(by=['사용 금액'], ascending=False)
category_amount
```

```python
category_amount.index.value_counts(normalize=True)
```

```python
category = pd.merge(category_count,category_amount, on="업종")
category = category.sort_values(by=['사용 금액'], ascending=False)
category
```

```python
category['사용 건수'].plot(kind="bar")
```

```python
category['사용 건수'].plot(kind="pie")
```

```python
category['사용 금액'].plot(kind="bar")
```

```python
category['사용 금액'].plot(kind="pie")
```

```python
df_category = category[0:5]
df_category
```

```python
# 사용금액 TOP 5
colors = ['#004358', '#FD7400', '#BF0917','#FFE11A', '#1F8A70']

plt.pie(df_category['사용 금액'], shadow=False, colors=colors, startangle=90)
plt.axis('equal')
plt.legend(df_category.index, loc='lower center', ncol=5, bbox_to_anchor=(0.5, -0.30))
plt.show
```
